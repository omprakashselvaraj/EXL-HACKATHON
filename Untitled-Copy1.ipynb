```python
# Replace 'scroll_pause_time' with the desired pause time between scrolls (in seconds)
scroll_pause_time = 1

# Replace 'scrolls' with the number of times you want to scroll to load more profiles
scrolls = 2

# Scroll multiple times to load more profiles
for _ in range(scrolls):
    linkedin.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(scroll_pause_time)
```

```python
src = linkedin.page_source
soup = BeautifulSoup(src, 'html.parser')
links = soup.find_all('a')
profile_links = [link['href'] for link in links if 'linkedin.com/in' in link['href']]
```

```python
links = list(set(profile_links))
```

```python
s = 'https://www.linkedin.com/in/ajay-2521'
s = s.split('?')
```

```python
cleaned = []
for i in links:
    i = i.split('?')
    cleaned.append(i[0])
```

```python
cleaned
```

```python
names = []
companies = []

for i in cleaned:
    fulllink = i
    linkedin.get(fulllink)
    time.sleep(8)
    src = linkedin.page_source
    soup = BeautifulSoup(src, 'lxml')

    #scraping name
    name = soup.find('title').text
    names.append(name)
    company = soup.find('div', {'class': 'text-body-medium break-words'}).text
    companies.append(company)
    
```

```python
import pandas as pd
df = pd.DataFrame(list(zip(names, companies)), 
               columns =['Name', 'Title'])
```
